```python
import requests
from bs4 import BeautifulSoup

# Déterminez le nombre total de pages
nombre_de_pages = 36

# Boucle pour parcourir chaque page
for numero_page in range(1, nombre_de_pages + 1):
    # Construire l'URL de la page avec le numéro de page actuel
    url = f"https://www.marmiton.org/recettes/recherche.aspx?aqt=cocktail&page={numero_page}"

    # Effectuer une requête HTTP pour récupérer le contenu HTML de la page
    response = requests.get(url)
    contenu_html = response.content

    # Utiliser BeautifulSoup pour analyser le contenu HTML
    soup = BeautifulSoup(contenu_html, 'html.parser')

    # Trouver tous les éléments <a> avec la classe "MRTN__sc-1gofnyi-2 gACiYG"
    liens = soup.find_all('a', class_="MRTN__sc-1gofnyi-2 gACiYG")
    for lien in liens:
        # Extraire et afficher le lien de chaque résultat
        lien_page = lien['href']
        print(f'https://www.marmiton.org{lien_page}')
```

### Code Rich

```python
import requests
from bs4 import BeautifulSoup

def retour_liste_recette(nom_recette):
    lower_nom_recette = nom_recette.lower()
    nom_recette_adapte = '-'.join(lower_nom_recette.split(" "))

    liste_liens = []
    page = 1
    while True:
        url_recettes = f"https://www.marmiton.org/recettes/recherche.aspx?aqt={nom_recette_adapte}&page={page}"
        rep_recettes = requests.get(url_recettes)

        soup = BeautifulSoup(rep_recettes.text, "html.parser")
        liste_a = soup.find_all("a", class_="MRTN__sc-1gofnyi-2 gACiYG")

        if len(liste_a) == 0:
            break

        for recette in liste_a:
            liste_liens.append(recette['href'])

        page += 1

    return liste_liens



def retour_liste_ingredients(liste_liens, numero_recette):
    lien_recette = liste_liens[numero_recette-1]
    url_recette = f"https://www.marmiton.org{lien_recette}"
    rep_recette = requests.get(url_recette)

    soup = BeautifulSoup(rep_recette.text, "html.parser")

    liste_span = soup.find_all("span", class_="SHRD__sc-10plygc-0 kWuxfa")
    liste_elements = []
    for span in liste_span:
        liste_elements.append(span.text)

    liste_span = soup.find_all("span", class_="SHRD__sc-10plygc-0 epviYI")
    liste_quantite = []
    for span in liste_span:
        liste_quantite.append(span.text)

    liste_ingredients = []
    for iteration in range(len(liste_elements)):
        element = liste_elements[iteration]
        quantite = liste_quantite[iteration]
        ingredient = f"{quantite} {element}"
        liste_ingredients.append(ingredient)

    return liste_ingredients


nom_recette = input("Quelle recette souhaitez-vous ? ")
liste_liens = retour_liste_recette(nom_recette)

if len(liste_liens) == 0:
    print("Aucune recette trouvée.")
else:
    print(f"Nombre de recettes trouvées : {len(liste_liens)}")

for index, lien in enumerate(liste_liens):
    texte = 'recette_'
    debut = lien.find(texte)
    fin = lien.find('.aspx')
    print(f"{index+1} : {lien[debut+len(texte):fin]}")

numero_recette = int(input("Quel numéro de recette souhaitez-vous ? "))
if numero_recette > 0 and numero_recette <= len(liste_liens):
    liste_ingredients = retour_liste_ingredients(liste_liens, numero_recette)
    print(f"\nVoici la liste d'ingrédients pour {lien[debut+len(texte):fin]} :")
    for ingredient in liste_ingredients:
        print(f"- {ingredient}")
else:
    print("Numéro de recette invalide.")
```

```python
def retour_liste_recette(nom_recette):
    lower_nom_recette = nom_recette.lower()
    nom_recette_adapte = '-'.join(lower_nom_recette.split(" "))
    
    url_recettes = f"https://www.marmiton.org/recettes/recherche.aspx?aqt={nom_recette_adapte}"
    rep_recettes = requests.get(url_recettes)

    soup = BeautifulSoup(rep_recettes.text, "html.parser")

    liste_div = soup.find("div", class_="SHRD__sc-dvq2vt-4 fAebFg")
    liste_a = liste_div.find_all("a")
    dernier_lien = liste_a[-1].get('href')
    egal = dernier_lien.rfind('=')
    nombre_de_pages = int(dernier_lien[egal+1:])

    liste_a = soup.find_all("a", class_="MRTN__sc-1gofnyi-2 gACiYG")

    liste_liens = []
    for recette in liste_a:
        liste_liens.append(recette['href'])

    for numero_page in range(2,nombre_de_pages+1):
        url_recettes = f"https://www.marmiton.org/recettes/recherche.aspx?aqt={nom_recette_adapte}&page={numero_page}"
        rep_recettes = requests.get(url_recettes)

        soup = BeautifulSoup(rep_recettes.text, "html.parser")

        liste_a = soup.find_all("a", class_="MRTN__sc-1gofnyi-2 gACiYG")

        for recette in liste_a:
            liste_liens.append(recette['href'])
    return liste_liens, nombre_de_pages

def retour_liste_ingredients(liste_liens, numero_recette):
    lien_recette = liste_liens[numero_recette-1]
    url_recette = f"https://www.marmiton.org{lien_recette}"
    rep_recette = requests.get(url_recette)

    soup = BeautifulSoup(rep_recette.text, "html.parser")

    liste_span = soup.find_all("span", class_="SHRD__sc-10plygc-0 kWuxfa")
    liste_elements = []
    for span in liste_span:
        liste_elements.append(span.text)

    liste_span = soup.find_all("span", class_="SHRD__sc-10plygc-0 epviYI")
    liste_quantite = []
    for span in liste_span:
        liste_quantite.append(span.text)

    liste_ingredients = []
    for iteration in range(len(liste_elements)):
        element = liste_elements[iteration]
        quantite = liste_quantite[iteration]
        ingredient = f"{quantite} {element}"
        liste_ingredients.append(ingredient)
    return liste_ingredients

import requests
from bs4 import BeautifulSoup

nom_recette = input("Quelle recette souhaitez-vous?")
liste_liens, nombre_de_pages = retour_liste_recette(nom_recette)
recettes_propre = []
for index,lien in enumerate(liste_liens):
    texte = 'recette_'
    debut = lien.find(texte)
    fin = lien.find('.aspx')
    recettes_propre.append(lien[debut+len(texte):fin])
    # print(f"{index+1} : {recettes_propre[index]}")

numero_recette = 1
while numero_recette > 0 and numero_recette <= nombre_de_pages*12:
    numero_recette = int(input(f"Il existe {nombre_de_pages*12} recettes pour {nom_recette}. \n Quel numéro de recette souhaitez-vous?"))
    choix_recette_ok = input(f"Vous avez choisi {recettes_propre[numero_recette-1]} ; cela vous convient-il? (o/n)").lower()
    if choix_recette_ok == "o":
        break

liste_ingredients = retour_liste_ingredients(liste_liens, numero_recette)
print(f"Voici la liste d'ingrédients pour {recettes_propre[numero_recette-1]}:")
for ingredient in liste_ingredients:
    print(ingredient)
```
